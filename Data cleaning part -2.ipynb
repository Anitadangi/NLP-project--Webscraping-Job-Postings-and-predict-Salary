### Categorising job titles further

```python
def clean_title(text):
    pat = re.compile('.*business\sintelligence.*')
    matches = re.match(pat, text)
    if matches:
        return 'business_intelligence'
    else:
        return None
```

```python
def clean_title(text):
    pat = re.compile('.*engineer.*')
    matches = re.match(pat, text)
    if matches:
        return 'data engineer'
    else:
        return ' '
```

```python
df['data_engineer'] = df.job_title.map(lambda x: clean_title(x))
```

```python
def clean_title2(text):
    pat = re.compile('.*consultan(t|cy).*')
    matches = re.match(pat, text)
    if matches:
        return 'consultant'
    else:
        return ' '
```

```python
df['business_intelligence'] = df.job_title.map(lambda x: clean_title(x))
```

```python
df['consultant']= df.job_title.map(lambda x: clean_title2(x))
```

```python
df['senior_role'] = df['job_title'].str.contains('manager') | df['job_title'].str.contains('lead') | df['job_title'].str.contains('senior') | df['job_title'].str.contains('director')
```
